## # Grouping Data
First finding means of data by category then breaking down data into groups using two axes


```python
grouped =df[['Protein', 'Carbs g', 'Fat g', 'Calories (kcal)']].groupby(df['Category'])
grouped
```

```python
grouped.mean()
```

```python
means =df['Protein'].groupby([df['Category'], df['Food (Per 100g)']]).mean()
means
```

```python
means.unstack()
```

```python
def peak_to_peak(arr):
    return arr.max() - arr.min()
grouped.agg(peak_to_peak)
```
